# **TRANSLATION**

```python
def translation(api_key, user_id, input_lang, output_lang, text):


    url_config = "https://meity-auth.ulcacontrib.org/ulca/apis/v0/model/getModelsPipeline"

    headers_config = {
        "Content-Type": "application/json",
        "ulcaApiKey": api_key,
        "userID": user_id
    }

    payload_config = {
        "pipelineTasks": [{
            "taskType": "translation",
            "config": {
                "language": {
                    "sourceLanguage": input_lang,
                    "targetLanguage": output_lang
                }
            }
        }],


        "pipelineRequestConfig": {
            "pipelineId": "64392f96daac500b55c543cd"
        }
    }


    response = requests.post(url_config, json=payload_config, headers=headers_config)

    if response.status_code == 200:
        response_data = response.json()
        print("Success:", response_data)
    else:
        print("Error:", response.status_code, response.text)

    compute_url = response_data['pipelineInferenceAPIEndPoint']['callbackUrl']
    header_name = response_data['pipelineInferenceAPIEndPoint']['inferenceApiKey']['name']
    header_value = response_data['pipelineInferenceAPIEndPoint']['inferenceApiKey']['value']
    payload_serviceID = response_data['pipelineResponseConfig'][0]['config'][0]['serviceId']
    payload_modelId = response_data['pipelineResponseConfig'][0]['config'][0]['modelId']


    url_compute = compute_url


    headers_compute = {
        header_name : header_value
    }


    payload_compute = {
        "pipelineTasks": [
            {
                "taskType": "translation",
                "config": {
                    "language": {
                        "sourceLanguage": input_lang,
                        "targetLanguage": output_lang
                    },
                    "serviceId": payload_serviceID,
                    "modelId": payload_modelId
                }
            }
        ],
        "inputData": {
            "input": [
                {
                    "source": text
                }
            ]
        }
    }



    response = requests.post(url_compute, json=payload_compute, headers=headers_compute)


    if response.status_code == 200:

        translated_text = response.json()
        print("Translated text:", translated_text)
    else:
        print("Error:", response.status_code, response.text)


    output_text = translated_text['pipelineResponse'][0]['output'][0]['target']


    return output_text
```

```python
input_lang = "hi"  #hindi
output_lang = "en"  #english

text = "मेरा नाम विहिर है और मैं भाषाावर्ष यूज कर रहा हूँ"

translation(UCLA_API_KEY, USER_ID, input_lang, output_lang, text)
```

# **TTS**

```python
def tts(api_key, user_id, input_lang, gender, text):
    url_config = "https://meity-auth.ulcacontrib.org/ulca/apis/v0/model/getModelsPipeline"

    headers_config = {
        "Content-Type": "application/json",
        "ulcaApiKey": api_key,
        "userID": user_id
    }

    payload_config = {
        "pipelineTasks": [{
            "taskType": "tts",
            "config": {
                    "language": {
                        "sourceLanguage": input_lang
                    }
                }
        }],


        "pipelineRequestConfig": {
            "pipelineId": "64392f96daac500b55c543cd"
        }
    }

    response = requests.post(url_config, json=payload_config, headers=headers_config)

    if response.status_code == 200:
        response_data = response.json()
        print("Success:", response_data)
    else:
        print("Error:", response.status_code, response.text)


    compute_url = response_data['pipelineInferenceAPIEndPoint']['callbackUrl']
    header_name = response_data['pipelineInferenceAPIEndPoint']['inferenceApiKey']['name']
    header_value = response_data['pipelineInferenceAPIEndPoint']['inferenceApiKey']['value']
    payload_serviceID = response_data['pipelineResponseConfig'][0]['config'][0]['serviceId']
    payload_modelId = response_data['pipelineResponseConfig'][0]['config'][0]['modelId']

    url_compute = compute_url


    headers_compute = {
        header_name : header_value
    }


    payload_compute = {
        "pipelineTasks": [
            {
                "taskType": "tts",
                "config": {
                    "language": {
                        "sourceLanguage": input_lang
                    },
                    "serviceId": payload_serviceID,
                    "gender": gender
                }
            }
        ],
        "inputData": {
            "input": [
                {
                    "source": text
                }
            ]
        }
    }

    response = requests.post(url_compute, json=payload_compute, headers=headers_compute)

    if response.status_code == 200:
        response_data = response.json()
        print("Success:", response_data)
    else:
        print("Error:", response.status_code, response.text)

    base64_data = response_data['pipelineResponse'][0]['audio'][0]['audioContent']

    return base64_data
```

```python
text = "मेरा नाम विहिर है और मैं भाषाावर्ष यूज कर रहा हूँ"
input_lang = "hi"   #hindi
gender = "male"

base64_format = tts(UCLA_API_KEY, USER_ID, input_lang, gender, text)
```

```python
import base64

# Your Base64 encoded MP3 data

# Decode the Base64 encoded data
mp3_data = base64.b64decode(base64_format)

# Path where you want to save the MP3 file
mp3_file_path = "output.mp3"

# Write the decoded data to a file
with open(mp3_file_path, "wb") as mp3_file:
    mp3_file.write(mp3_data)

print(f"MP3 file has been saved to: {mp3_file_path}")
```

```python
def asr(api_key, user_id, input_lang, base64_input, audio_format, sampling_rate):

    url_config = "https://meity-auth.ulcacontrib.org/ulca/apis/v0/model/getModelsPipeline"

    headers_config = {
        "Content-Type": "application/json",
        "ulcaApiKey": UCLA_API_KEY,
        "userID": USER_ID
    }

    payload_config = {
        "pipelineTasks": [
            {
                "taskType": "asr",
                "config": {
                    "language": {
                        "sourceLanguage": input_lang
                    }
                }
            }],


        "pipelineRequestConfig": {
            "pipelineId": "64392f96daac500b55c543cd"
        }
    }


    response = requests.post(url_config, json=payload_config, headers=headers_config)

    if response.status_code == 200:
        response_data = response.json()
        print("Success:", response_data)
    else:
        print("Error:", response.status_code, response.text)


    compute_url = response_data['pipelineInferenceAPIEndPoint']['callbackUrl']
    header_name = response_data['pipelineInferenceAPIEndPoint']['inferenceApiKey']['name']
    header_value = response_data['pipelineInferenceAPIEndPoint']['inferenceApiKey']['value']
    payload_serviceID = response_data['pipelineResponseConfig'][0]['config'][0]['serviceId']
    payload_modelId = response_data['pipelineResponseConfig'][0]['config'][0]['modelId']


    url_compute = compute_url
    null = "null"

    headers_compute = {
        header_name : header_value
    }



    payload_compute = {
        "pipelineTasks": [
            {
                "taskType": "asr",
                "config": {
                    "language": {
                        "sourceLanguage": input_lang
                    },
                    "serviceId": payload_serviceID,
                    "audioFormat": audio_format,
                    "samplingRate": sampling_rate
                }
            }
        ],
        "inputData": {
            "input": [
                {
                    "source": null
                }
            ],
            "audio": [
                {
                    "audioContent": base64_input
                }
            ]
        }
    }


    # audioContent input in base64 only


    response = requests.post(url_compute, json=payload_compute, headers=headers_compute)

    if response.status_code == 200:
        asr = response.json()
        print("Translated text:", asr)
    else:
        print("Error:", response.status_code, response.text)


    asr_output = asr['pipelineResponse'][0]['output'][0]['source']

    return asr_output
```
